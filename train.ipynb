4bit quantization config

```python
from transformers import BitsAndBytesConfig

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
)
```

LoRA config

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=32,
    lora_alpha=64,
    lora_dropout=0.05,
    target_modules=["q_proj", "k_proj", "v_proj"],
    bias="all",
    task_type="CAUSAL_LM"
)
```

Tokenizer

```python
model_name = "mistralai/Mistral-Small-24B-Instruct-2501"
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
# A pad token should be added for the tokenizer
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

```python
def add_labels(example):
    tokenized = tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=512
    )
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized
```

Data collator config

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
    pad_to_multiple_of=8
)
```

Training

```python
from accelerate import notebook_launcher, Accelerator
from transformers import TrainingArguments, Trainer
from datasets import load_from_disk


def training_function():
    accelerator = Accelerator(
        mixed_precision='fp16',
        gradient_accumulation_steps=4
    )
    
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16,
        low_cpu_mem_usage=True,
        quantization_config=quant_config,
        device_map={"": accelerator.process_index}
    )
    
    model = get_peft_model(model, lora_config)
    model.print_trainable_parameters()

    dataset = load_from_disk("tokenized_mistral_dataset")
    dataset = dataset.map(add_labels, batched=True)
    dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])

    dataset = dataset.train_test_split(test_size=0.1)
    train_dataset = dataset['train'].shard(num_shards=accelerator.num_processes, index=accelerator.process_index)
    valid_dataset = dataset['test'].shard(num_shards=accelerator.num_processes, index=accelerator.process_index)
    
    print(dataset['train'][0]['labels'][:10])
    print(type(dataset['train'][0]['labels']))
    
    training_args = TrainingArguments(
        output_dir="./mistral-24b-clm-qlora",
        per_device_train_batch_size=8,
        gradient_accumulation_steps=4,
        eval_strategy="steps",
        eval_steps=400,
        save_strategy="steps",
        save_steps=400,
        save_total_limit=2,
        logging_steps=100,
        num_train_epochs=2,
        learning_rate=2e-4,
        fp16=True,
        optim="paged_adamw_8bit",
        warmup_steps=200,
        lr_scheduler_type="cosine",
        report_to="wandb",
        run_name="mistral-lora-run",
        load_best_model_at_end=True,
        metric_for_best_model="eval_loss",
        greater_is_better=False,
        dataloader_num_workers=4,
        ddp_backend="nccl",
        ddp_find_unused_parameters=False,
        deepspeed=None,
        remove_unused_columns=False
    )
    
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=valid_dataset,
        tokenizer=tokenizer,
        data_collator=data_collator
    )
    
    accelerator.print(f"Num GPUs: {accelerator.num_processes}")
    trainer.train()
```

```python
notebook_launcher(training_function, num_processes=2)
```
